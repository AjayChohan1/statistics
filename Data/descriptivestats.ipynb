```python
import pandas as pd

# Create a sample dataset similar to retail sales
data = {
    'Product': ['Shirt', 'Pants', 'Shoes', 'Shirt', 'Shoes', 'Pants', 'Shirt', 'Shoes', 'Pants', 'Shirt'],
    'Category': ['Clothing', 'Clothing', 'Footwear', 'Clothing', 'Footwear', 'Clothing', 'Clothing', 'Footwear', 'Clothing', 'Clothing'],
    'Price': [25.99, 35.50, 60.00, 22.50, 65.00, 40.00, 20.00, 55.00, 38.00, 23.99],
    'Units_Sold': [120, 80, 60, 150, 50, 90, 130, 65, 85, 140]
}

df = pd.DataFrame(data)
df.head(15)  # Display the first 15 rows of the DataFrame
# Calculate total sales for each product
```

```python
df.describe()
# Calculate total sales for each product
df['Total_Sales'] = df['Price'] * df['Units_Sold']
```

```python
df['Category'].value_counts()
# Calculate total sales by category
total_sales_by_category = df.groupby('Category')['Total_Sales'].sum().reset_index()
```

```python
df['Price'].mean()       # Average price
df['Units_Sold'].median() # Median units sold
df['Price'].std()        # Standard deviation of prices
# Display the total sales by category
print(total_sales_by_category)
```

```python
df.groupby('Category')['Price'].mean()
# Display the average price by category
average_price_by_category = df.groupby('Category')['Price'].mean().reset_index()
```

```python
df.groupby('Category')['Price'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='Category', y='Price', data=df)
plt.title("Price Distribution by Category")
plt.show()
# Display the average price by category
print(average_price_by_category)
```
